```python
import pandas as pd
import numpy as np
```

```python
# read the csv file
df = pd.read_csv('players_20.csv')

# set index
df.set_index('short_name', inplace=True)

# select columns
df = df[['long_name', 'age', 'dob', 'height_cm', 'weight_kg', 'nationality', 'club']]

#display the data frame
df.head()
```

```python
# update value of a cell
df.loc['L. Messi', 'height_cm'] = 175
#df.loc['L. Messi', 'height_cm']
df['height_cm'].head(3)
```

```python
# update values of multiple cells
df.loc[['L. Messi', 'Cristiano Ronaldo'], ['height_cm']] = 165
df['height_cm'].head(3)
```

```python
# update values in a column
df.loc[:, 'age'] = 30
df.loc[df.index[:5], 'age']
```

```python
# update values in a row
df.loc[df.index[-1], :] = np.nan
print(df.iloc[-1], '\n')

df.iloc[-1,:] = 0
df.tail(1)
```

```python
# set value for rows matching a condition
df.loc[df['height_cm']>180, 'age'] = 27
df.head()
```
