```python
import pandas as pd
```

```python
# read the csv file
df = pd.read_csv('players_20.csv')

# set index
df.set_index('short_name', inplace=True)

# select columns
df = df[['long_name', 'age', 'dob', 'height_cm', 'weight_kg', 'nationality', 'club']]

#display the data frame
df.head()
```

```python
# drop one row
df.drop('L. Messi').head(3)
```

```python
# index parameter
df.drop(index=['Cristiano Ronaldo']).head(3)
```

```python
# drop two or more rows and update data (inplace=True)
df.drop(['J. Oblak', 'Neymar Jr'], inplace=True)
df.head(3)
```

```python
# drop one column
df.drop('long_name', axis=1).head(3)
```

```python
# columns parameter
df.drop(columns=['long_name']).head(3)
```

```python
# drop column by position/index (drop last column)
df.drop(df.columns[[-1]], axis=1).head(3)
```

```python
# drop two or more columns and update data (inplace=True)
df.drop(columns=['weight_kg', 'age'], inplace=True)
df.head(3)
```
