```python
import pandas as pd
```

```python
# read the csv file and display the data frame
df = pd.read_csv('laptop_price.csv')
df.head(3)
```

## Filter a Data Frame Based on One Condition 

```python
# find which rows have "Apple" in the "Company" column
condition_apple = (df['Company']=='Apple')
condition_apple[:3]
```

```python
# filter dataframe based on condition
apple_laptops = df[condition_apple]
apple_laptops.head(3)
```

```python
# verify which companies are in the "Company" column with value_counts()
df[condition_apple].value_counts('Company')
```

```python
# find which rows don't have "HP" in the "Company" column
condition_not_hp = df['Company']!='HP'
condition_not_hp[:3]
```

```python
not_hp_laptops = df[condition_not_hp].value_counts('Company')
not_hp_laptops
```

```python
# find laptops with price over 2000 euros
df[df['Price_euros']>2000].head()
```

## Filter a Data Frame Based on Multiple Conditions 

```python
# apple laptops which costs more than 2000 euros - using &(and)
apple_expensive = (df['Company'] == 'Apple') & (df['Price_euros'] > 2000)
df[apple_expensive]
```

```python
# apple or dell laptops - using |(or)
apple_or_dell = (df['Company'] == 'Apple') | (df['Company'] == 'Dell')
df[apple_or_dell]['Company'].value_counts()
```

```python
# apple or dell laptops which cost more than a 2000 euros - using | and &Â¶
apple_dell_expensive = ((df['Company'] == 'Apple') | (df['Company'] == 'Dell')) & (df['Price_euros'] > 2000)
df[apple_dell_expensive].head()
```
