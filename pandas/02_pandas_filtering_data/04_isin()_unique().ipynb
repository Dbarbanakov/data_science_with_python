```python
import pandas as pd
```

```python
# read the csv file and display the data frame
df = pd.read_csv('laptop_price.csv')
df.head(3)
```

## .isin() 

```python
# single filtering
# select Apple or HP Laptops
filter1 = df['Company'].isin(['Apple', 'HP'])
df[filter1].head(3)
```

```python
df[filter1]['Company'].value_counts()
```

```python
# multiple filtering
# find notebooks or ultabooks from Apple or HP
filter2 = df['TypeName'].isin(['Notebook', 'Ultrabook'])
df[filter1 & filter2].head(3)
```

```python
df[filter1 & filter2]['Company'].value_counts()
```

## .unique()

```python
# get unique elements in the "Company" column
check_uniques = df['Company'].unique()
check_uniques
```

```python
len(check_uniques)
```

## .nunique() 

```python
# get the number of unique elements in a column
count_uniques = df['Company'].nunique()
count_uniques
```
