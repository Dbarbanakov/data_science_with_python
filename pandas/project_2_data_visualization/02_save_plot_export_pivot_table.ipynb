```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# read the csv file
df_raw = pd.read_csv('population_total.csv')

# first look at the dataset
df_raw.head()
```

## Create a Pivot Table

.pivot(): Returns reshaped DataFrame organized by given index / column values ("pivot without aggregation")

```python
# drop null values
df_raw.dropna(inplace=True)
```

```python
# create a pivot table
df_pivot = df_raw.pivot(index='year', columns='country',
                                        values='population')
```

```python
# select some countries
df_pivot = df_pivot[['United States', 'India', 'China', 
                          'Indonesia', 'Brazil']]
```

```python
# show pivot table
df_pivot
```

```python
# export pivot table
df_pivot.to_excel('pivot_table.xlsx')
```
