```python
import pandas as pd
```

```python
# read the csv file and display the data frame
df = pd.read_csv('laptop_price.csv')
df.head(3)
```

## Find Duplicate Rows

### .duplicated() 

```python
# find duplicates in 1 column/series
duplicates = df['laptop_ID'].duplicated()
df[duplicates]
```

```python
duplicates.value_counts()
```

```python
# duplicates in 2 or more columns
more_duplicates = df.duplicated(['Product', 'TypeName', 'Inches'])
more_duplicates.value_counts()
```

```python
# show all the values duplicated (except for the first - keep='first')
df[more_duplicates].sort_values(['Product', 'Inches'])
```

Task: Find the cheapest and most expensive laptop of each company using the sort_values() and duplicated() methods (keep "first", "last" and False)

```python
# sort dataframe asending by company and price
df = df.sort_values(['Company', 'Price_euros'])
df.head()
```

```python
# check all value categories in "Company" column
df.value_counts(['Company'])
```

```python
# duplicated values in "Company" columns (default -> keep='first
duplicated_first = df.duplicated(['Company'])

# show dataframe with duplicated values in "Company" column
df[duplicated_first].head()
```

```python
# keep='first'(cheapest laptop per company)
# show dataframe with non duplicated values in "Company" column
df[~duplicated_first]
df[~duplicated_first][['Company', 'Price_euros']]
```

```python
# check all value categories
df[~duplicated_first].value_counts('Company')
```

```python
# keep='last'
duplicated_last = df.duplicated('Company', keep='last')
# keep=False
duplicated_false = df.duplicated('Company', keep=False)
```

```python
# keep='last'(most expensive laptop per company)
df[~duplicated_last]
# show dataframe with non duplicated values in "Company" column
df[~duplicated_last][['Company', 'Price_euros']]
```

```python
# keep=False
# show dataframe with non duplicated values in "Company" column
df[~duplicated_false]
```

```python
# all elements in company column are repeated at least once
df.value_counts(['Company'])
```

## Drop Duplicate Elements 

### drop_duplicates() 

```python
# drop duplicates in 2 or more columns
df.drop_duplicates(['Company'])
df.drop_duplicates(['Company'])[['Company', 'Price_euros']]
```

```python
# sort dataframe asending by company and price
# cheapest first and most expensive last
df = df.sort_values(['Company', 'Price_euros'])
df.head()
```

```python
# cheapest: keep='first'
df.drop_duplicates(['Company'], keep='first')[['Company', 'Price_euros']]
```

```python
# most expensive: keep='last'
df.drop_duplicates(['Company'], keep='last')[['Company', 'Price_euros']]
```
