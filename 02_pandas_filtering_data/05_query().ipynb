```python
import pandas as pd
```

```python
# read the csv file and display the data frame
df_laptops = pd.read_csv('laptop_price.csv')
df_laptops.head(3)
```

```python
# select laptops based on one condition

# boolean slicing
df_laptops[df_laptops['Inches'] > 15.6]

# query method
df_laptops.query('Inches > 15.6').head(3)
```

```python
# add a not operator in the upper examples

# boolean slicing
df_laptops[~(df_laptops['Inches'] > 15.6)]

# query method
df_laptops.query('not Inches > 15.6').head(3)
```

```python
# select laptops based on multiple conditions
ram_condition = df_laptops['Ram'].str.extract('(\d+)', expand=False).astype(int) > 16
price_condition = df_laptops['Price_euros'] < 2000

# boolean slicing
df_laptops[ram_condition & price_condition]

# query method
df_laptops.query('Price_euros < 2000 and @ram_condition') 
```
