```python
import pandas as pd
```

```python
# read the csv file and display the data frame
df_exams = pd.read_csv('StudentsPerformance.csv')
df_exams.head(3)
```

## Column Operations

```python
print(f'sum - {df_exams['math score'].sum()}')
print(f'count - {df_exams['math score'].count()}')
print(f'mean - {df_exams['math score'].mean()}')
print(f'std - {df_exams['math score'].std()}')
print(f'min - {df_exams['math score'].min()}')
print(f'max - {df_exams['math score'].max()}')
```

```python
# easier calculation with the describe method
df_exams.describe()
```

## Row Operations

```python
# calculate the total sum of a row
rows_sum = df_exams['math score'] + df_exams['reading score'] + df_exams['writing score']
rows_sum.head(3)
```

```python
# rounding to 2 decimals the average score and assigning the result to a new column
df_exams['average score'] = (rows_sum / 3).round(2)
df_exams.head(3)
```
