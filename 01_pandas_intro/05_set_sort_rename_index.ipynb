```python
import pandas as pd
import numpy as np
import random
```

```python
# read the csv file and display the data frame
df_exams = pd.read_csv('StudentsPerformance.csv')
df_exams.head(3)
```

### Create Index

```python
# create unique values for the index column
new_index = np.arange(0,1000)
new_index[:5]
```

```python
# shuffle the index array
random.shuffle(new_index)
new_index[:5]
```

```python
# create a column with new_index
df_exams['new index'] = new_index
df_exams.head(3)
```

```python
# set new_index column as index
df_exams.set_index('new index', inplace=True)
df_exams.head(3)
```

```python
# sort the Data Frame by the index column
df_exams.sort_index(inplace=True)
df_exams.head(3)
```

### Rename Index

```python
# rename index 0:A, 1:B, 2:C and update the Data Frame
df_exams.rename(index={0:'A', 1:'B', 2:'C'}, inplace=True)
df_exams.head(4)
```
